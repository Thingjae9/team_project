```python
import time
import tensorflow as tf
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from keras.callbacks import EarlyStopping
from sklearn.preprocessing import StandardScaler
from keras.layers import Dropout
from tensorflow.keras.layers import BatchNormalization
```

```python
df = pd.read_csv('csv/reg_standard.csv')
target = 'Rings'
y = df[target]
x = df.drop(target, axis =1)
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size = 0.2, random_state=42)
# method_custom_metric 구현
def accuracy(y_true, y_pred):
    return 1 - tf.abs((y_true - y_pred) / y_true) 
```

```python

# 모델 구현
model = Sequential()
model.add(Dense(64, activation='tanh', input_dim=x.shape[1]))
model.add(Dense(32, activation='tanh'))              
model.add(Dense(16, activation='tanh'))
model.add(Dense(16, activation='tanh'))
model.add(Dense(1, activation='linear'))

# 모델 컴파일
model.compile(loss='mse', optimizer='adam', metrics=[accuracy])

# early stopping 구현 - 커스텀 정확도 기준
early_stopping = EarlyStopping(monitor='val_accuracy', patience=25)
start_time = time.time()
model.fit(X_train, y_train, epochs=1000, batch_size=64, validation_data=(X_val, y_val), callbacks=[early_stopping])
end_time = time.time()
long_time = end_time - start_time
train_loss, train_acc = model.evaluate(X_train, y_train, verbose =2)
val_loss, val_acc = model.evaluate(X_test, y_test, verbose=2)



```

```python
func = 'tanh'
opti = 'adam'
batch = 128
def accuracy(y_true, y_pred):
    return 1 - tf.abs((y_true - y_pred) / y_true) 

model = Sequential()
model.add(Dense(24, activation=func, input_dim=x.shape[1], kernel_initializer = 'he_normal'))
model.add(BatchNormalization())  # BatchNormalization 추가
model.add(Dense(24, activation=func, kernel_initializer = 'he_normal'))
model.add(BatchNormalization())  # BatchNormalization 추가
model.add(Dropout(0.1))
model.add(Dense(32, activation=func, kernel_initializer = 'he_normal'))
model.add(BatchNormalization())  # BatchNormalization 추가
model.add(Dropout(0.1))
model.add(Dense(64, activation=func, kernel_initializer = 'he_normal'))
model.add(BatchNormalization())  # BatchNormalization 추가
model.add(Dropout(0.1))
model.add(Dense(22, activation=func, kernel_initializer = 'he_normal'))
model.add(BatchNormalization())  # BatchNormalization 추가
model.add(Dropout(0.1))
model.add(Dense(12, activation=func, kernel_initializer = 'he_normal'))
model.add(BatchNormalization())  # BatchNormalization 추가
model.add(Dropout(0.1))
model.add(Dense(5, activation=func, kernel_initializer = 'he_normal'))
model.add(BatchNormalization())  # BatchNormalization 추가
model.add(Dropout(0.1))
model.add(Dense(1, activation='linear'))

# 모델 컴파일
model.compile(loss='mse', optimizer=opti, metrics=[accuracy])

# early stopping 구현 - 커스텀 정확도 기준
early_stopping = EarlyStopping(monitor='accuracy', patience=20)
start_time = time.time()
model.fit(X_train, y_train, epochs=1000, batch_size=batch, validation_data=(X_test, y_test), callbacks=[early_stopping])
end_time = time.time()
long_time = end_time - start_time
loss, acc = model.evaluate(X_val, y_val, verbose=2)
```
